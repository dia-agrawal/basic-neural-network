```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
print(X_train.shape, y_train.shape)
X_train = X_train.reshape(-1, 28*28).astype('float32') / 255
X_test = X_test.reshape(-1, 28*28).astype('float32') / 255
X_train.shape, y_train.shape
# y_train=keras.utils.to_categorical(y_train)#,num_classes=)
# y_test=keras.utils.to_categorical(y_test)
```

```python
# bmi
height = np.round(np.random.normal(1.75, 0.20, 10000), 2)
weight = np.round(np.random.normal(60.32, 15, 10000), 2)
np_city = np.column_stack((height, weight))
bmi = weight/height**2 
```

```python
X_train, X_test, y_train, y_test = train_test_split(np_city, bmi, test_size=0.2, random_state=42)
X_train[0], y_train[0]
```

```python
X_train[1], y_train[1]
```

```python
# binary_step_function
# parameterised_relu
activation_fn = ['linear', 'tanh', 'relu', 'leaky_relu', 'elu', 'swish', 'sigmoid', 'softmax']

# model = tf.keras.Sequential([
#     tf.keras.layers.Dense(units = 128, activation = activation_fn[2], input_shape = (784,)),
#     tf.keras.layers.Dropout(0.2), # WHY we set ranodm nodes to zero?
#     tf.keras.layers.Dense(units = 10, activation = 'softmax')
# ])
# model.summary()

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 32, activation = activation_fn[2], input_shape = (2,)), # 128 , 64 what is the difference?
    tf.keras.layers.Dense(units = 8, activation = activation_fn[2]),
    tf.keras.layers.Dense(units = 2, activation = activation_fn[2]),
    tf.keras.layers.Dense(units = 1, activation = 'relu')
])
model.summary()
```

```python
from keras import backend as K
from tensorflow.keras.metrics import MeanAbsoluteError, MeanSquaredError, RootMeanSquaredError

def r2_score(y_true, y_pred):
    SS_res =  K.sum(K.square( y_true-y_pred ))
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) )
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )


optimizers = ['Adadelta', 'Adagrad', 'Adam', 'SGD', 'RMSprop']  # have to read this
loss_fn = ['mean_squared_error',  'mean_absolute_error',  'binary_crossentropy',  'categorical_crossentropy', 'sparse_categorical_crossentropy']
    # 'huber',
    # 'kullback_leibler_divergence',  
    # 'poisson',  
    # 'cosine_similarity' ]

model.compile(optimizer = optimizers[2],
            loss = loss_fn[0],
             metrics = [r2_score])
```

```python
model.fit(X_train, y_train, epochs =200)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test) 
```
